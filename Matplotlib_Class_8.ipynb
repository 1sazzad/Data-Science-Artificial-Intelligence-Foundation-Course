## Pyplot submodule and plt alias
Most of the Matplotlib utilities lies under the `pyplot` submodule, and are usually imported under the `plt` alias.
Using `plt` alias we can referred pyplot packages instead of using `matplotlib.pyplot`.



```
# import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = [2023, 2024, 2025, 2026]
y = [15, 25, 30, 20]

# x = np.array([2023, 2024, 2025, 2026])
# y = np.array([15, 25, 30, 20])
plt.plot(x, y) # The plot function is used to draw points(markers) in a diagram.
plt.show()
```

## **Application** of `**dict_name`

The `**` used to unpack the dictionary

if $D = \{ k_1 : v_1, k_2 : v_2, \dots, k_n : v_n \}$

$f(**D) = f(k_1 = v_1, k_2 = v_2, \dots, k_n = v_n)$

**Conceptual Diagram:**



```
line_style (dict)
        ↓
** unpacking
        ↓
converted to keyword arguments
        ↓
received in plot() as kwargs
        ↓
Matplotlib applies styling
```

### Let's do something funny

````python
import matplotlib.pyplot as plt
import numpy as np

H = 10
W = 6
gap = 4

x_points = []
y_points = []

for y in range(H):
    for x in range(3*W + 2*gap):

        # --- LETTER I ---
        if x < W:
            if y == 0 or y == H-1 or x == W//2:
                x_points.append(x)
                y_points.append(H - y)

        # --- LETTER S ---
        elif W + gap <= x < 2*W + gap:
            xs = x - (W + gap)

            if (
                y == 0 or
                y == H//2 or
                y == H-1 or
                (y < H//2 and xs == 0) or
                (y > H//2 and xs == W-1)
            ):
                x_points.append(x)
                y_points.append(H - y)

        # --- LETTER T ---
        elif x >= 2*W + 2*gap:
            xt = x - (2*W + 2*gap)

            if y == 0 or xt == W//2:
                x_points.append(x)
                y_points.append(H - y)

plt.figure(figsize=(8,4))
plt.scatter(x_points, y_points)
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.title("IST plotted mathematically")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-1.5, 1.5, 0.01)
y = np.arange(-1.5, 1.5, 0.01)
X, Y = np.meshgrid(x, y)

F = (X**2 + Y**2 - 1)**3 - X**2 * Y**3
crack = (X > 0) & (Y < 0.5*X) & (Y > 0.5*X - 0.2)

mask = (F <= 0) & (~crack)

plt.figure(figsize=(6,6))
plt.scatter(X[mask], Y[mask], c='#E24B5A')
plt.gca().set_aspect('equal')
plt.axis("off")
plt.show()
````

## Labels
```
Figure
│
└── Axes
    │
    └── Axis (X-axis, Y-axis)
        │
        └── Text (labels, ticks, title)
```


```
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([20, 44, 35, 30])
y3 = np.array([10, 24, 30, 10])

label_format = dict(
    fontsize = 12,
    family = "serif",
    fontweight = "bold",
    color = "#2d4cfc"
)
plt.title("Class Size", **label_format)
plt.xlabel("Year", **label_format)
plt.ylabel("Student", **label_format)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xticks(x)
plt.show()
```
