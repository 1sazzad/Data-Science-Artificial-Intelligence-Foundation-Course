## **Application** of `**dict_name`

The `**` used to unpack the dictionary

if $D = \{ k_1 : v_1, k_2 : v_2, \dots, k_n : v_n \}$

$f(**D) = f(k_1 = v_1, k_2 = v_2, \dots, k_n = v_n)$

**Conceptual Diagram:**



```
line_style (dict)
        ↓
** unpacking
        ↓
converted to keyword arguments
        ↓
received in plot() as kwargs
        ↓
Matplotlib applies styling
```

### Let's do something funny

````python
import matplotlib.pyplot as plt
import numpy as np

H = 10
W = 6
gap = 4

x_points = []
y_points = []

for y in range(H):
    for x in range(3*W + 2*gap):

        # --- LETTER I ---
        if x < W:
            if y == 0 or y == H-1 or x == W//2:
                x_points.append(x)
                y_points.append(H - y)

        # --- LETTER S ---
        elif W + gap <= x < 2*W + gap:
            xs = x - (W + gap)

            if (
                y == 0 or
                y == H//2 or
                y == H-1 or
                (y < H//2 and xs == 0) or
                (y > H//2 and xs == W-1)
            ):
                x_points.append(x)
                y_points.append(H - y)

        # --- LETTER T ---
        elif x >= 2*W + 2*gap:
            xt = x - (2*W + 2*gap)

            if y == 0 or xt == W//2:
                x_points.append(x)
                y_points.append(H - y)

plt.figure(figsize=(8,4))
plt.scatter(x_points, y_points)
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.title("IST plotted mathematically")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-1.5, 1.5, 0.01)
y = np.arange(-1.5, 1.5, 0.01)
X, Y = np.meshgrid(x, y)

F = (X**2 + Y**2 - 1)**3 - X**2 * Y**3
crack = (X > 0) & (Y < 0.5*X) & (Y > 0.5*X - 0.2)

mask = (F <= 0) & (~crack)

plt.figure(figsize=(6,6))
plt.scatter(X[mask], Y[mask], c='#E24B5A')
plt.gca().set_aspect('equal')
plt.axis("off")
plt.show()
````

````python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([20, 44, 35, 30])
y3 = np.array([10, 24, 30, 10])

label_format = dict(
    fontsize = 12,
    family = "serif",
    fontweight = "bold",
    color = "#2d4cfc"
)
plt.title("Class Size", **label_format)
plt.xlabel("Year", **label_format)
plt.ylabel("Student", **label_format)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xticks(x) #
# plt.grid()
plt.show()
````

## Sub Plots


````python
import matplotlib.pyplot as plt
import numpy as np

# Create simple arrays
x = np.array([1, 2, 3, 4, 5])

y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])
y3 = np.array([5, 3, 4, 2, 1])
y4 = np.array([10, 8, 6, 4, 2])

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title("Array 1")

axs[0, 1].plot(x, y2)
axs[0, 1].set_title("Array 2")

axs[1, 0].plot(x, y3)
axs[1, 0].set_title("Array 3")

axs[1, 1].plot(x, y4)
axs[1, 1].set_title("Array 4")

plt.tight_layout()
plt.show()
````

# Task
**Goal:** Visualize marks of 5 students in 4 subjects.

**Requirement**


1.   bar()
2.   subplot()
3.  legend()

**Data Example**

```
students = ["A", "B", "C", "D", "E"]

math = np.array([80, 75, 90, 85, 70])
physics = np.array([78, 72, 88, 80, 68])
chemistry = np.array([85, 70, 92, 88, 75])
english = np.array([90, 85, 95, 89, 80])
```
**What you do must:**

2. Each subplot &rarr; one subject
3. Use bar chart
4. Add titles, labels
5. Add subtitles

**Skills you will Practice:**

````python
import matplotlib.pyplot as plt
import numpy as np

students = ["A", "B", "C", "D", "E"]

math = np.array([80, 75, 90, 85, 70])
physics = np.array([78, 72, 88, 80, 68])
chemistry = np.array([85, 70, 92, 88, 75])
english = np.array([90, 85, 95, 89, 80])


fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].bar(students, math)
axs[0, 0].set_title("Math")
axs[0, 0].set_xlabel("Students")
axs[0, 0].set_ylabel("Marks")

axs[0, 1].bar(students, physics)
axs[0, 1].set_title("Physics")
axs[0, 1].set_xlabel("Students")
axs[0, 1].set_ylabel("Marks")

axs[1, 0].bar(students, chemistry)
axs[1, 0].set_title("Chemistry")
axs[1, 0].set_xlabel("Students")
axs[1, 0].set_ylabel("Marks")


axs[1, 1].bar(students, english)
axs[1, 1].set_title("English")
axs[1, 1].set_xlabel("Students")
axs[1, 1].set_ylabel("Marks")

plt.suptitle("Marks of 5 Students in 4 Subjects")
plt.tight_layout()
plt.show()
````
